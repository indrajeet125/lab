```python
import numpy as np
from pylab import *
```

```python
def N(x):
    condi1=x<0
    condi2=np.logical_and(0<=x,x<1)
    condi3=np.logical_and(1<=x,x<2)
    condi4=x>=2
    r=np.where(condi1,0,0)
    r=np.where(condi2,x,r)
    r=np.where(condi3,2-x,r)
    r=np.where(condi4,0,r)
    return r
x=np.array([1,-1,2,4,5,6,-6,7,4,0,1,1.1,2])
y=N(x)
plot(x,y)
```
