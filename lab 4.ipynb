# Second Question

```python
def f(x):
    return x*(12-x)+np.sin(np.pi*x)


def midpoint(f, a, b, n):
    h = (b-a)/n
    res = 0
    i = 1
    while i <= n:
        res += f((a-0.5*h)+i*h)
        i += 1
    res *= h
    return res


def vectormidpoint(f, a, b, n):
    result = 0
    if isinstance(n, int):
        n=np.arange(1,n+1,1)
    h=(b-a)/len(n)
        
    result=f(a-0.5*h+n*h)
    result=np.sum(result)*h
    return result


res = midpoint(f, 0, np.pi, 10)
print('Result using loop=', res)


res = vectormidpoint(f, 0, np.pi, 10)
print('Result using vectorised=', res)
arr=np.arange(1,11)
res = vectormidpoint(f, 0, np.pi,arr)
print('Result-2 using vectorised=', res)
plt.plot(res)
plt.show()


```

```python
def s(x, n):
    ans = 0
    for j in range(n+1):
        ans += ((-1)**j)*(x**(2*j+1))/(math.factorial(2*j+1))

    return ans
n=5
while(n<=16):
    no_points=10
    while(no_points<=100):
        x = np.linspace(0, np.pi*4, no_points)
        y = np.array([s(xt, n) for xt in x])
        plt.title(f"n={n} no.points{no_points}")
        plt.plot(x, y)
        plt.show()
        no_points+=20
    n+=3
```
